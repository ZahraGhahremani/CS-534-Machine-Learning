We can do this through singular value decomposition $X = USV$, where V are orincipal axis, S is a diagonal matrix with singular values, and U are principal components scaled to unit norm. Standardized PCs van be calculated by $N-1\sqrt{U}$. PCs themselves (also known as "scores") are given by columns of $US$.Theoretically now that covariance matrix is given by $\frac{1}{N-1}X^TX=V\frac{s^2}{N-1}V^T=V\sqrt{E}$, so principal axes V are eigenvectors of the covariance matrix and $E=\frac{s^2}{N-1}$ are its eigenvalues.
Now the cross-covariance matrix (or loadings matrix: it is given by the eigenvectors of the covariance matrix scaled by the square roots of the respective eigenvalues.) between original variables and standardized PCs $$r=ν_{(ij)}\frac{std(Y_{j})}{std(X_{i})}$$where $ν_{ij}$ is an i-th element of the j-th unit-length eigenvector of the covariance matrix, $var(Y_{j})$ is the corresponding eigenvalue which gives variance of this PC, and $std(X_{i})$ is standard deviation of $X_{i}$.

## answering the question:
 Creating functions for calculating the gini index, split a dataset, and entropy.



```python
def splitt(index, value, data):
  left, right = list(), list()
  for row in data:
    if row[index] < value:
      left.append(row)
    else:
      right.append(row)
  return left, right


def Gini(groups, classes):
    n_instances = float(sum([len(group) for group in groups]))
    gini = 0.0
    for group in groups:
        size = float(len(group))
        if size == 0:
            continue
        score = 0.0
        for class_val in classes:
            p = [row[-1] for row in group].count(class_val) / size
            score += p * p
        gini += (1.0 - score) * (size / n_instances)
    return gini

def Entropy(groups, classes):
    n_instances = float(sum([len(group) for group in groups]))
    entropy = 0.0
    for group in groups:
        size = float(len(group))
        if size == 0:
            continue
        score = 0.0
        for class_val in classes:
            p = [row[-1] for row in group].count(class_val) / size
            score += p * np.log(p)
        entropy += (- score) * (size / n_instances)
    return entropy
```

Creating a function to return the group maximum and consider that as a terminal node value

```python
def Terminal(group):
    outcomes = [row[-1] for row in group]
    return max(set(outcomes), key=outcomes.count)
```

Crating a split function to produce child splits for a node or make terminal and another function to choose the best split point for a dataset.



```python
def split(node, max_depth, min_size, depth, Metric):
    left, right = node['groups']
    del(node['groups'])
    if not left or not right:
        node['left'] = node['right'] = Terminal(left + right)
        return
    if depth >= max_depth:
        node['left'], node['right'] = Terminal(left), Terminal(right)
        return
    if len(left) <= min_size:
        node['left'] = Terminal(left)
    else:
        node['left'] = Find_best_split(left,Metric)
        split(node['left'], max_depth, min_size, depth+1, Metric)
    if len(right) <= min_size:
        node['right'] = Terminal(right)
    else:
        node['right'] = Find_best_split(right,Metric)
        split(node['right'], max_depth, min_size, depth+1, Metric)
        
def Find_best_split(data,Metric):
    class_values = list(set(row[-1] for row in data))

    b_index, b_value, b_score, b_groups = 999, 999, 999, None
    for index in range(len(data[0])-1):
        for row in data:
            groups = splitt(index, row[index], data)            
            Metric_index = Metric(groups, class_values)

            if Metric_index < b_score:
                b_index, b_value, b_score, b_groups = index, row[index], Metric_index, groups
    return {'index':b_index, 'value':b_value, 'groups':b_groups}
```

Decision tree



```python
def decision_tree(train, max_depth, min_size,Metric):
  root = Find_best_split(train,Metric)
  split(root, max_depth, min_size, 1, Metric)
  def illustrator(root, depth=0):
    if isinstance(root, dict):

      header = 'rank'
      if root['index']+1 == 1:
        header = 'GRE';
      if root['index']+1 == 2:
        header = 'GPA';

      print('%s[%s < %.3f]' % ((depth*' ', (header), root['value'])))
      illustrator(root['left'], depth+1)
      illustrator(root['right'], depth+1)
    else:
      print('%s[%s]' % ((depth*' ', root)))
  illustrator(root)
  return root
```

## Answering:
loading the UCLA dataset of admissions (only the first 30 samples)

```python
Admission = pd.read_csv("/content/drive/MyDrive/Spring 2022-MS-Semester2/ML/HW1/binary.csv")
# link to the dataset: https://stats.idre.ucla.edu/stat/data/binary.csv
change = Admission.pop('admit')
Admission['admit']=change
Admission = Admission.iloc[0:30]
Admission.head()
```

Creating train and test dataset

```python
train = Admission.sample(frac=0.8)
test_mask = pd.Series(True, index=Admission.index)
test_mask[train.index] = False
test = Admission[test_mask]
test.head()
```

Train the tree using Gini ratio

```python
Train_tree = decision_tree(train.values.tolist(), 3, 5, Gini)
```

Creating a function to predict based on our tree

```python
def predictor(node, row):
    if row[node['index']] < node['value']:
        if isinstance(node['left'], dict):
            return predictor(node['left'], row)
        else:
            return node['left']
    else:
        if isinstance(node['right'], dict):
            return predictor(node['right'], row)
        else:
            return node['right']
```

Checking our trained tree

```python
Train_results = pd.DataFrame(columns=['decision', 'admit', 'counter'])
i=0
for row in train.values.tolist():
    prediction = predictor(Train_tree, row)
    Train_results.loc[i]=[prediction,row[3],i]
    i+=1
    print('True_observation=%d, Prediction_tree=%d' % (row[-1], prediction))
#Train_results.head()
```
